Гипотеза о том что сотрудники с высоким окладом проводят на работе больше времени чем с низким

```python
df_low=HR[HR['salary'] == 'low']['average_montly_hours']
df_low
```

```python
df_h=HR[HR['salary'] == 'high']['average_montly_hours']
df_h
```

```python
tvalue, pvalue = stats.ttest_ind(df_low, df_h)
print('t_value:',tvalue, '_pvalue: ',pvalue)
```

Мы получаем хорошее значение p, равное  0.4661 и, таким образом, мы отклоняем нулевую гипотезу, то есть - средние значения двух 
распределений различны и это различие статистически значимо
