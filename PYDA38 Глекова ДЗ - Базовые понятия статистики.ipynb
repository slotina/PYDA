```python
q1 = df3['Age'].quantile(0.25)
q3 = df3['Age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df3[df3['Age'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

```python
df3[~df3['Age'].between(lower_bound, upper_bound, inclusive=True)]
```
